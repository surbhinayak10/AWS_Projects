```python
# Importing all the required libraries
import pandas as pd
import boto3
from io import StringIO
```

* **Loading Data into pandas**

```python
df = pd.read_csv('https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/mushrooms.csv')
# Displaying first 5 rows of dataset
df.head()
```

* **Creating Bucket in S3**

```python
s3_client = boto3.client('s3')
s3_client.create_bucket(Bucket='iansurbhi')
```

* **Upload file to s3 bucket**

```python
bucket = 'iansurbhi' # already created on S3
csv_buffer = StringIO()
df.to_csv(csv_buffer)
s3_resource = boto3.resource('s3')
s3_resource.Object(bucket, 'mushroom.csv').put(Body=csv_buffer.getvalue())
```

* **Output a confirmation that file is on s3 bucket**

```python
bucket = s3.Bucket('iansurbhi')

for obj in bucket.objects.all():
    key = obj.key
    body = obj.get()['Body'].read()
    print(key)
    print(body)
```
